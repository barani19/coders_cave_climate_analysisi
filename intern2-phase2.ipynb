```python
daily_data.drop(columns=['STATION','REPORT_TYPE','SOURCE','BackupElements','BackupElevation','BackupEquipment','BackupName','WindEquipmentChangeDate'], inplace=True)
daily_data.head()
```

```python
from datetime import datetime
daily_data['DATE'] = pd.to_datetime(daily_data['DATE'],infer_datetime_format=True)
daily_data.head()
```

```python
daily_data.info()
```

```python
select = ['DailyAverageDryBulbTemperature', 'DailyPrecipitation', 'DailyAverageWindSpeed'] 
select_val = daily_data[select].describe()
select_val.describe()
```

```python
plt.plot(daily_data['DATE'], daily_data['DailyAverageDryBulbTemperature'], color='blue')
plt.title('Time Series Plot for Daily Average Temperature')
plt.xlabel('Date')
plt.ylabel('Temperature (°F)')
plt.show()

plt.plot(daily_data['DATE'], daily_data['DailyPrecipitation'], color='green')
plt.title('Time Series Plot for Daily Precipitation')
plt.xlabel('Date')
plt.ylabel('Precipitation (inches)')
plt.show()

plt.plot(daily_data['DATE'], daily_data['DailyAverageWindSpeed'], color='orange')
plt.title('Time Series Plot for Daily Average Wind Speed')
plt.xlabel('Date')
plt.ylabel('Wind Speed (mph)')
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decompose = seasonal_decompose(daily_data['DailyAverageDryBulbTemperature'], model='additive', period=30)
decompose.plot()
```
